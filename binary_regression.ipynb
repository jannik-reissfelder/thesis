```python
# define helper fuction for MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    """
    Takes as input y_true and y_pred and calculates the MAPE error between both vectors. Note that y_true and y_pred are both ndarray numpy arrays.
    """
    # Avoid division by zero and handle the infinity
    y_true = np.where(y_true == 0, np.finfo(float).eps, y_true)

    # Compute the absolute percentage error
    ape = np.abs((y_true - y_pred) / y_true)

    # Calculate the MAPE
    mape = np.mean(ape) * 100  # Multiply by 100 to get the percentage

    return mape, ape
```

```python
def cross_validation(X_df, y_df, cv, model):
        """
        cross validation
        :return:

        """
        mse_scores = []
        rmse_scores = []
        mape_scores = []
        r2_scores = []

        classes = y_df.index
        # extract values from dataframe
        X_df = X_df.values
        y_df = y_df.values

        # Loop over the folds
        i = 0
        # for train_index, test_index in cv.split(X_df, classes):
        for train_index, test_index in cv.split(X_df):
            X_train, X_test = X_df[train_index], X_df[test_index]
            y_train, y_test = y_df[train_index], y_df[test_index]
            i += 1
            print(f"""Fitting Fold {i}""")
            # Fit the model
            model.fit(X_train, y_train)

            # Make predictions
            y_pred = model.predict(X_test)

            # Calculate the mean squared error for the current fold
            mse = mean_squared_error(y_test, y_pred)
            mse_scores.append(mse)
            rmse = np.sqrt(mse)
            rmse_scores.append(rmse)


            # Calculate the mean absolute percentage error for the current fold
            mape, ape = mean_absolute_percentage_error(y_test, y_pred)
            mape_scores.append(mape)


            # Calculate R2 score for each output separately
            score = r2_score(y_test, y_pred)

            # Append the average R2 score to the list
            r2_scores.append(score)
        
        return np.mean(mse_scores), np.mean(rmse_scores), np.mean(mape_scores), np.mean(r2_scores)
```

```python
targets = Y.columns.tolist()
```

```python
print("some")
mse_scores_dict = {}
rmse_scores_dict = {}
mape_scores_dict = {}
r2_scores_dict = {}
i = 0
for target in targets:
    i += 1
    print(f"""Fitting target {i}""")
    mse_scores, rmse_scores, mape_scores, r2_scores = cross_validation(X, Y[target], cv, model)
    mse_scores_dict[target] = mse_scores
    rmse_scores_dict[target] = rmse_scores
    mape_scores_dict[target] = mape_scores
    r2_scores_dict[target] = r2_scores

print("FFoo")
```

```python
# convert to dataframe
mse_scores_df = pd.DataFrame.from_dict(mse_scores_dict, orient='index', columns=['mse'])
rmse_scores_df = pd.DataFrame.from_dict(rmse_scores_dict, orient='index', columns=['rmse'])
mape_scores_df = pd.DataFrame.from_dict(mape_scores_dict, orient='index', columns=['mape'])
r2_scores_df = pd.DataFrame.from_dict(r2_scores_dict, orient='index', columns=['r2'])

# merge dataframes
scores_df = pd.concat([mse_scores_df, rmse_scores_df, mape_scores_df, r2_scores_df], axis=1)
```
